```python
for index, row in cost_of_liv.iterrows():
    city_name = cost_of_liv.loc[index, "city"]
    target_url = f"https://api.openweathermap.org/geo/1.0/direct?q={city_name},US&limit=5&appid={api_key}"
    city_data = requests.get(target_url).json()
    
    try:
        print("start state")
        cost_of_liv.loc[index, "state"]  = city_data[0]["state"]
    except(KeyError, IndexError):
        # If no city is found, set the state to no state found".
        cost_of_liv.loc[index, "state"] = "No state found"

print("done looping")
```
