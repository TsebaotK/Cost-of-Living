```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

cost_of_liv_exp_info = pd.read_csv("output_expense_breakdown_states")
States = cost_of_liv_exp_info["state"].unique()
Expenses = cost_of_liv_exp_info["Expenses_outside_CC"].mean()
group_by_state = cost_of_liv_exp_info.groupby(["state"]).mean(["Expenses_outside_CC"]).sort_values(["Expenses_outside_CC"], ascending = False)
Expenses
```

```python
#Finding Regional Trends by state

States
Expense = group_by_state["Expenses_outside_CC"]

plt.figure(figsize = (15,5))
plt.bar(group_by_state.index, Expense, color="r", alpha = 0.5, width = 0.6)
plt.title("Trending Expense by State")
plt.xlabel("State")
plt.ylabel("Average Mpnthly Expense")
plt.xticks(rotation = 90)
plt.savefig("Trend_by_region.png", bbox_inches="tight")
plt.show()




```
