```python
import pandas as pd
import numpy as np
```

```python
cost_of_liv = pd.read_csv("cost-of-living_v2.csv")
```

```python
cost_of_liv
```

```python
cost_of_liv_US = cost_of_liv.loc[cost_of_liv["country"] == "United States"]
cost_of_liv_US
```

```python
# Grocery list include eggs, bread, milk, meat, beef, apples, bananas, orannges, tomatoes, potatoes, onions, lettuce
grocery_df = cost_of_liv_US[["city","x9","x10","x11","x12","x13", "x14", "x15", "x16", "x17", "x18", "x19", "x20", "x21",
                            "x22","x23"]]

#grocery_selected_df = grocery_df.rename(columns = {"city":"City", "x9":"milk","x10":"bread", "x11":"rice","x12":"eggs",
#                                                   "x13":"cheese", "x14":"chicken", "x15":"beef",
#                                                   "x16":"apple", "x17":"banana", "x18":"orange",
#                                                  "x19":"tomato", "x20":"potato", "x21":"onion",
#                                                   "x22":"lettuce","x23":"water-1.5L bottle"})
total_grocery = grocery_df["x9"] + grocery_df["x10"] + grocery_df["x11"] + grocery_df["x12"] + grocery_df["x13"] + grocery_df["x14"] + grocery_df["x15"] + grocery_df["x16"]+ grocery_df["x17"] + grocery_df["x18"] + grocery_df["x19"] + grocery_df["x20"]+ grocery_df["x21"] + grocery_df["x22"] + grocery_df["x23"]
#final_ grocery = grocery_selected_df.groupby(["City"])
grocery_df["Total_Grocery_Monthly"] = total_grocery
#grocery_df
#grocery_selected_df.loc[grocery_selected_df["City"]].sum()

final_grocery_df = grocery_df[["city", "Total_Grocery_Monthly"]]
final_grocery_df
```

```python
cost_US_df = cost_of_liv_US[["city", "country", "x29","x33", "x36", "x38", "x50", "x51", "x54"]]
cost_US_renamed_df = cost_US_df.rename(columns = {"city":"city", "country":"Country", "x29":"Monthly Transp Pass", "x33":"Gas/liter", "x36":"Utilities",
                                             "x38":"Internet", "x50":"3 bed aptt in City Center", "x51":"3 bed aptt outside Center",
                                             "x54":"Avg. Monthly Salary"})

cost_US_renamed_df
```

```python
complete_df = pd.merge(cost_US_renamed_df, final_grocery_df, on = "city")
complete_df
```

```python
# Cleaning the "Nan" /empty rows:

complete_cleaned_df = complete_df.dropna(how = "any")
complete_cleaned_df
```

```python
#Finding the monthly Gas expense (Multiplying by 4(for 4 times in a month), 60 liters is average a car holds)
complete_cleaned_df["Gas/liter"] *= 4 * 60

#Finding the monthly grocery Multiplying by 4(for 4 times in a month)
complete_cleaned_df["Total_Grocery_Monthly"] *= 4
# Finding the expenses, given that the accomadation is outside the city center
complete_cleaned_df["Expenses_outside_CC"] = complete_cleaned_df["Monthly Transp Pass"] + (complete_cleaned_df["Gas/liter"]) + complete_cleaned_df["Utilities"] + complete_cleaned_df["Internet"] + complete_cleaned_df["3 bed aptt outside Center"] + complete_cleaned_df["Total_Grocery_Monthly"]
complete_cleaned_df
```

```python
#Finding the correlation between the cost of living(Expenses_outside_CC) and the Average monthly Salary; 
#Lower the ratio, more affordable, and higher the ratio, the more expensive.
complete_cleaned_df["Expense/Income"] = complete_cleaned_df["Expenses_outside_CC"]/ complete_cleaned_df["Avg. Monthly Salary"]
complete_cleaned_df = round(complete_cleaned_df, 2)
complete_cleaned_df
```
